#EDA

>analyzing the relationship between cooking sessions and user orders

```python
data['Session_Hour'] = data['Session_Start'].dt.hour
```

```python
data['Session_Duration'] = (data['Session_End'] - data['Session_Start']).dt.total_seconds() / 60

session_orders = data.groupby('Session_Hour').agg({
    'Order_ID': 'count',
    'Session_Duration': 'mean'
}).reset_index()
session_orders.rename(columns={'Order_ID': 'Total_Orders'}, inplace=True)
correlation = session_orders[['Session_Duration', 'Total_Orders']].corr()
correlation
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap: Session Duration vs Total Orders")
plt.show()
```

* Positive Relationship:

A longer session duration is moderately associated with a higher number of orders. This suggests that when cooking sessions are longer, more orders are likely being fulfilled. However, the correlation is not very strong, so other factors might also influence the total number of orders.

* Practical Implications:

If the goal is to maximize orders, it might be beneficial to analyze what makes sessions longer and optimize processes to handle more orders efficiently during those extended periods.
Investigate whether longer sessions are planned (e.g., for peak hours or special events) or due to delays, and how this impacts order efficiency.

```python
time_slot_orders = data.groupby('Session_Hour').agg({'Order_ID': 'count'}).reset_index()
time_slot_orders.rename(columns={'Order_ID': 'Order_Count'}, inplace=True)

plt.figure(figsize=(10, 6))
sns.barplot(x='Session_Hour', y='Order_Count', data=time_slot_orders, palette='viridis')
plt.title('Orders by Session Time of Day', fontsize=16)
plt.xlabel('Session Hour (24-hour format)', fontsize=12)
plt.ylabel('Order Count', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

* Popular food

```python
popular_dishes = data.groupby('Dish_Name').agg({'Order_ID': 'count'}).reset_index()
popular_dishes.rename(columns={'Order_ID': 'Order_Count'}, inplace=True)
# Sort by order count to find the most popular dishes
popular_dishes_sorted = popular_dishes.sort_values(by='Order_Count', ascending=False)
popular_dishes_sorted
```

```python
top_popular_dishes = popular_dishes_sorted.head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(y='Order_Count', x='Dish_Name', data=top_popular_dishes, palette='viridis')
plt.title('Top 10 Popular Dishes', fontsize=16)
plt.xlabel('Dish Name', fontsize=12)
plt.ylabel('Order Count', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

>Examining Demographic Trends

```python
demographic_orders = data.groupby('Age').agg({'Order_ID': 'count'}).reset_index()
demographic_orders.rename(columns={'Order_ID': 'Total_Orders'}, inplace=True)
demographic_orders['Age_Bin'] = pd.cut(demographic_orders['Age'],bins = [24,29,34,39,44],labels = ['25-30','30-35','35-40','40-45'] )
```

```python
age_bin_orders = demographic_orders.groupby('Age_Bin')['Total_Orders'].sum()

age_bin_orders.plot.pie(
    autopct='%1.1f%%',
    figsize=(8, 8),
    cmap='tab10',
    title='Proportion of Total Orders by Age Group'
)
plt.ylabel('')
plt.show()
```

```python
loc_data = data.groupby('Location')['Order_ID'].count().sort_values(ascending = False).reset_index()
loc_data.rename(columns={'Order_ID': 'Total_Orders'}, inplace=True)
loc_data
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(y='Total_Orders', x='Location', data=loc_data, palette='viridis')
plt.title('Orders by Location', fontsize=16)
plt.xlabel('Location', fontsize=12)
plt.ylabel('Order Count', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
fav_data = data.groupby('Favorite_Meal').agg({'Order_ID':'count'}).reset_index()
fav_data.rename(columns={'Order_ID':'Order_Count'},inplace = True)
fav_data
```

# Summary of Findings

1. Correlation Between Session Duration and Total Orders

*  The correlation coefficient between session duration and total orders is 0.453.

* This moderate positive correlation suggests that longer session durations are associated with a higher number of orders, but other factors may also influence order counts.

2. Orders by Session Time of Day

* Peak hours for orders are 7:00 PM (5 orders) and 6:00 PM (3 orders).

* Afternoon sessions (12:00 PM, 1:00 PM) also show moderate activity.

* Morning sessions (7:00 AM, 8:00 AM) have minimal activity.

This suggests that evening time slots are the most popular for sessions and should be the focus for maximizing engagement and order volume.

3. Popular Dishes

* Grilled Chicken (4 orders) > Spaghetti (4 orders) > Caesar Salad (3 orders)

* Dishes like Pancakes, Veggie Burger, and Oatmeal have lower popularity.

The presence of "No Session/Orders" indicates missing or unlinked data points, which should be addressed for improved accuracy.

4. Demographics and Order Patterns

* Age Group Analysis:

  * The majority of orders come from the 25-30 age group (38.9%).

  * The 30-35 and 35-40 age groups contribute equally at 22.2% each.

  * The 40-45 age group contributes the least at 16.7%.

* Younger audiences (25-30) are the most active demographic, suggesting targeted marketing strategies for this group could enhance engagement.

5. Orders by Location

* Locations with the highest number of orders:

  * Chicago, Los Angeles, and New York each contributed 3 orders.

  * San Francisco and Seattle contributed 2 orders each.

* Other cities (Austin, Boston, Dallas, Miami, Phoenix) had lower activity with 1 order each.

* Major cities like Chicago, Los Angeles, and New York are key markets for driving orders.

6. Favorite Meals

* Meal preferences ranked by order count:

    * Dinner is the most popular meal, contributing 8 orders.

    * Lunch follows with 6 orders.

    * Breakfast is the least popular meal with 4 orders.

    * Dinner is the prime opportunity for engagement and promotions.
